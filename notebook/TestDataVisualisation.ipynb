```julia
using Plots
```

```julia
data = [1 2 0;
        1.5 2.5 0;
        3 4 1; 
        1 5 0; 
        1 3 1; 
        3.5 5 1]

labels = data[:, 3]
data = data[:, 1:2]

blue = data[vec(labels .== 0), :]
red = data[vec(labels .== 1), :]

green = [0 0; 4 4] # new points

scatter(blue[:, 1], blue[:, 2], mc = "blue", ylimits=(-1,6),xlimits=(-1,6), label="class 0")
scatter!(red[:, 1], red[:, 2], mc = "red", label="class 1")
scatter!(green[:, 1], green[:, 2], mc = "green", label="new points")
```
